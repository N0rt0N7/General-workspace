```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('2024-05-07-task113-1v.txt')  

x = data[:, 0]  
y = data[:, 1]  

plt.figure(figsize=(10, 5))

y_R = np.divide(y, 27)
y_Rlog = np.log(y_R)

ionization = 15.1900
idx = np.abs(x - ionization).argmin()

plt.axvline(x=ionization, color='red', linestyle='--', label=f'X={ionization}')
plt.text(ionization, 0.000007, f'Ionization energy is {x[idx]:.2f}eV', color='black', fontsize=10, ha='center', va='top', 
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='square,pad=0.3'))


plt.plot(x, y_R)
plt.xlabel('Voltage - $U_G$')
plt.ylabel('Logarithm of Current Log(I)')
plt.title('Measurements per 1V')
plt.grid(True)
plt.show()
```
