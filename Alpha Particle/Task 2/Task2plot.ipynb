```python
import numpy as np
import matplotlib.pyplot as plt

y = []

def read_file_with_skipped_lines(file_path, num_lines_to_skip_first, num_lines_to_skip_last):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        # Skip the first x lines
        lines = lines[num_lines_to_skip_first:]
        # Skip the last y lines
        lines = lines[:-num_lines_to_skip_last]
        for line in lines:
            y.append(int(line.strip()))  # Process the remaining lines
            
            

x = list(range(1, 4095))
    
# Example usage
file_path = "task2-11mm-vacuum-45min.txt"  # Replace with the actual file path
num_lines_to_skip_first = 13
num_lines_to_skip_last = 15
read_file_with_skipped_lines(file_path, num_lines_to_skip_first, num_lines_to_skip_last)

y_max = max(y)
print("Peak at: ",y_max)

index_y_max = y.index(y_max)

plt.axvline(x=index_y_max, color='red', linestyle='-.')
plt.xlim([2800, 3500])
plt.plot(x, y)
plt.xlabel('Energy (keV)')
plt.ylabel('Intencity (a.u.)') 
plt.title('Plot for distance of 11mm from detector in vacuum at 45 min')
plt.grid(True)
#plt.show()
```
