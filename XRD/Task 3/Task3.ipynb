```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

file_path = 'powder.csv'
powder_data = pd.read_csv(file_path, skiprows=21)

reference_peaks = [29.720, 31.650, 31.751, 33.789, 34.930, 35.830, 36.160, 37.910, 46.449, 47.490, 48.020, 56.358, 60.298, 61.978, 62.209, 63.522, 66.091, 67.351, 68.619, 68.808, 71.080, 71.328, 73.509, 76.721, 81.048, 88.837]

plt.figure(figsize=(10, 6))
plt.plot(powder_data['Angle'], powder_data['Intensity'], label='Intensity')

for rp in reference_peaks:
    plt.axvline(x=rp, color='blue', linestyle='--', label='Zinc Indium Gallium Oxide' if rp == reference_peaks[0] else "")

plt.xlabel('Angle (2θ)')
plt.ylabel('Intensity')
plt.legend()
plt.xlim(10,90)
plt.grid(True)
plt.show()
```
