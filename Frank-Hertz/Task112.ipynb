```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('2024-05-07-task112-RT-06V.txt')  

x = data[:, 0]  
y = data[:, 1]  

y_R = np.divide(y, 27)
y_Rlog = np.log(y_R)

plt.plot(x, y_Rlog)
plt.xlabel('Voltage - V')
plt.ylabel('Logarithm of Current Log(I)')
plt.title('6V - RT')
plt.grid(True)
plt.show()
```
