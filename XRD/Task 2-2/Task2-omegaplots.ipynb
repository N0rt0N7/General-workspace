```python
import pandas as pd
from matplotlib import pyplot as plt
import plotly.express as px
  
plt.figure(figsize=(8,4))

df = pd.read_csv('ZnO-peak 125.28 (006) w-scan.csv', skiprows=21)
x = df['Angle']
y = df['Intensity']

#peak_list = [28.420, 47.306, 56.103, 69.173, 76.372, 88.059, 94.975, 106.735, 114.130, 127.574, 136.933]
#peak_orientation = ['(111)', '(220)', '(311)', '(400)', '(331)', '(422)', '(511)', '(440)', '(531)', '(620)', '(533)']

plt.xlim([60, 65])
plt.yscale('log')

plt.xlabel('ω Angle')
plt.ylabel('Intensity')

#for position in peak_list:
    #plt.axvline(x=position, color='red', linestyle='--', label=f'X={position}')
    #orient = peak_list.index(position)
    #plt.text(position, 0.8, f'{peak_orientation[orient]}', color='black', fontsize=9, ha='center', bbox=dict(facecolor='white', edgecolor='black', boxstyle='square,pad=0.2'))

plt.title('XRD of ZnO (a-plane)')
plt.grid(True) 
plt.plot(x,y,linewidth=0.5)
```
