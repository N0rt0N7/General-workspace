```python
import numpy as np
import matplotlib.pyplot as plt


data = np.loadtxt('2024-05-07-task122-txtfile.txt')  

x = data[:, 0]  
y = data[:, 1]  


Hg_spectral_lines_x = [253.62, 296.49, 365.08, 404.78, 435.95, 546.25]
artifact_lines = [312.86, 578.79]

plt.figure(figsize=(12.5, 5), dpi=100)

indices = np.searchsorted(x, Hg_spectral_lines_x)
Hg_spectral_lines_y = y[indices]

plt.xlabel('Wavelenght (nm)')
plt.ylabel('Value')
plt.xlim(178.80,886.38)


for position in Hg_spectral_lines_x:
    plt.axvline(x=position, color='red', linestyle='--', label=f'X={position}')
    plt.text(position, 0.5, f'{position}', color='black', fontsize=8, ha='center', bbox=dict(facecolor='white', edgecolor='black', boxstyle='square,pad=0.2'))

for position in artifact_lines:
    plt.axvline(x=position, color='green', linestyle='--', label=f'X={position}')
    plt.text(position, 0.5, f'{position}', color='green', fontsize=8, ha='left', bbox=dict(facecolor='white', edgecolor='black', boxstyle='square,pad=0.2'))

#print("Extracted y-values:", Hg_spectral_lines_y)


plt.grid(False)
plt.plot(x, y)
```
