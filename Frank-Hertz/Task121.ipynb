```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

data = np.loadtxt('2024-05-07-task121.txt')  

x = data[:, 0]  
y = data[:, 1]  

plt.figure(figsize=(10, 6))

peaks, _ = find_peaks(y, prominence=0.1)
sorted_peaks = sorted(peaks, key=lambda idx: -y[idx])
first_5_peaks = sorted_peaks[:5]



plt.semilogy(x, y)
plt.xlabel('Voltage - V')
plt.ylabel('Logarithm of Current Log(I)')
plt.title('Measurements per Task 121')
plt.plot(x[first_5_peaks], y[first_5_peaks], 'D', label='Peaks', color='red')  
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

data = np.loadtxt('2024-05-07-task121-attempt2.txt')  

x = data[:, 0]  
y = data[:, 1]  

peaks, _ = find_peaks(y, prominence=0.1)
sorted_peaks = sorted(peaks, key=lambda idx: -y[idx])
first_5_peaks = sorted_peaks[:5]

plt.figure(figsize=(8, 6))

plt.plot(x, np.log(y))
plt.xlabel('Voltage - V')
plt.ylabel('Logarithm of Current Log(I)')
plt.title('Measurements per Task 121')
plt.plot(x[first_5_peaks], y[first_5_peaks], 'D', label='Peaks', color='red')  
plt.grid(True)
plt.show()
```
